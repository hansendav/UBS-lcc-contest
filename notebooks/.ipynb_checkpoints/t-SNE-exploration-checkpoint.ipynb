```python
import os 
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
patches_path = '../dataset/training_patches'
labels_path = '../dataset/training_noisy_labels'

patches = []

for filename in os.listdir(patches_path):
    if filename.endswith('.png'):
        # Open the image file
        img = Image.open(os.path.join(patches_path, filename))
        
        # Convert the image to a numpy array and get the band values
        patches.append(np.array(img))

# Convert the list of band values to a numpy array
patches = np.array(patches)
```

```python
labels = []

for filename in os.listdir(labels_path):
    if filename.endswith('.png'):
        # Open the image file
        img = Image.open(os.path.join(labels_path, filename))
        
        # Convert the image to a numpy array and get the band values
        labels.append(np.array(img))

# Convert the list of band values to a numpy array
labels = np.array(labels)
```

```python
print(patches.shape)
print(labels.shape)
```

```python
labels_3ch = np.repeat(labels[..., np.newaxis], 3, axis=3)
```

```python
masked_patches = np.ma.masked_where(labels_3ch == 0, patches)
```

```python
filled_patches = masked_patches.filled(0)
```

```python
# Calculate histograms for each channel - label
hist_red_labels = np.histogram(filled_patches[..., 0], bins=256, range=(1, 256))[0]
hist_green_labels = np.histogram(filled_patches[..., 1], bins=256, range=(1, 256))[0]
hist_blue_labels = np.histogram(filled_patches[..., 2], bins=256, range=(1, 256))[0]

# Calculate histograms for each channel - full
hist_red_patch = np.histogram(patches[..., 0], bins=256, range=(1, 256))[0]
hist_green_patch = np.histogram(patches[..., 1], bins=256, range=(1, 256))[0]
hist_blue_patch = np.histogram(patches[..., 2], bins=256, range=(1, 256))[0]

# Plot the histograms using subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

axs[0].plot(hist_red, color='red')
axs[0].plot(hist_red_patch, color='red', linestyle='dashed')    
axs[0].set_title('Red Channel')
axs[0].set_xlabel('Pixel value')
axs[0].set_ylabel('Frequency')

axs[1].plot(hist_green, color='green')
axs[1].plot(hist_green_patch, color='green', linestyle='dashed')
axs[1].set_title('Green Channel ')
axs[1].set_xlabel('Pixel value')
axs[1].set_ylabel('Frequency')

axs[2].plot(hist_blue, color='blue')
axs[2].plot(hist_blue_patch, color='blue', linestyle='dashed')
axs[2].set_title('Blue Channel')
axs[2].set_xlabel('Pixel value')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
